```python
import pandas as pd
```

```python
melb_data = pd.read_csv('data/melb_data_fe.csv')
melb_data.head()
```

```python
melb_df = melb_data.copy()
```

```python
melb_df.info()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
quarter = melb_df['Date'].dt.quarter
quarter.value_counts()
```

```python
unique_list = []
for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtype)
    unique_list.append(item)
unique_counts = pd.DataFrame(
    data= unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
).sort_values(by='Num_Unique',  ignore_index=True)
display(unique_counts)
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_counts = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_counts and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

```python
melb_df.sort_values(by='AreaRatio', ignore_index=True).iloc[1557]['BuildingArea']
```

```python
melb_df[(melb_df['Type']== 'townhouse') & (melb_df['Rooms'] > 2)].sort_values(by=['Rooms','MeanRoomsSquare'], ascending=[True,False], ignore_index=True).iloc[18]['Price']
```

```python
melb_df.groupby(by='Type').mean(numeric_only=True)
```

```python
melb_df.groupby('Type')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
```

```python
melb_df.groupby('Rooms')['Price'].mean().sort_values()
```

```python
melb_df.groupby(['Regionname'])['Lattitude'].std(numeric_only=True).sort_values()
```
