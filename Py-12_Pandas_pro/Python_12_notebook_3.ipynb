```python
import pandas as pd
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv')
```

```python
ratings_movies
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'].info()
```

```python
# 8.2
movie_1999 = ratings_movies[ratings_movies['year_release'] == 1999]
movie_1999.groupby('title')['rating'].mean().sort_values()
```

```python
movie_2010 = ratings_movies[ratings_movies['year_release'] == 2010]
movie_2010.groupby('genres')['rating'].mean().sort_values()
```

```python
ratings_movies.groupby(['userId'])['genres'].agg(set).apply(len).sort_values()
```

```python
ratings_movies.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'],ascending=[True,False])
```

```python
ratings_movies[ratings_movies['year_release'] == 2018].groupby('genres')['rating'].agg(['mean', 'count']).sort_values(by=['mean', 'count'], ascending=[False,True])
```

```python
ratings_movies['date'] = pd.to_datetime(ratings_movies['date'])
ratings_movies['year_rating'] = ratings_movies['date'].dt.year
pivot_1 = pd.pivot_table(ratings_movies, values='rating', aggfunc='mean', index='genres', columns='year_rating' , fill_value=0)
pivot_1
```
